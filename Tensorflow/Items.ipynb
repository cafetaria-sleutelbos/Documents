```python
import pandas
import matplotlib.pyplot as plt
from collections import Counter
```

```python
items = pandas.read_csv("items.csv")
items.columns = ['item_id', 'name', 'cooking_time', 'complexity', 'price', 'created_all']
print(items)
```

```python
items.describe()
```

```python
ct_distinctX = Counter(items['cooking_time']).keys() # equals to list(set(words))
ct_distinctY = Counter(items['cooking_time']).values() # counts the elements' frequency
```

```python
plt.bar(x=ct_distinctX, height=ct_distinctY, align='center', width=25)
plt.xlabel("Cooking time (seconds)")
plt.ylabel("Amount of items")
plt.show()
```

```python
c_distinctX = Counter(items['complexity']).keys() # equals to list(set(words))
c_distinctY = Counter(items['complexity']).values() # counts the elements' frequency
```

```python
plt.bar(x=c_distinctX, height=c_distinctY, align='center', width=0.5)
plt.xlabel("Cooking time (seconds)")
plt.ylabel("Amount of items")
plt.show()
```
